```python
import numpy as np
import matplotlib.pyplot as plt

A = 1
T = 1
fs = 10
t = np.linspace(0, 2*T, 2*T*fs)
x = A * np.sin(2 * np.pi * t)


x_quantized_1bit = np.sign(x)


levels_2bit = [-1.5, -0.5, 0.5, 1.5]
x_quantized_2bit = np.digitize(x, levels_2bit) - 3


binary_1bit = ['0' if val == -1 else '1' for val in x_quantized_1bit]
binary_2bit = [format(val, '02b') if val >= 0 else format(val & 0b11, '02b') for val in x_quantized_2bit]


plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(t, x, label='Continuous Signal')
plt.title('Continuous Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(3, 1, 2)
plt.step(t, x_quantized_1bit, where='mid', label='One-bit Quantized')
plt.title('One-bit Quantization')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(3, 1, 3)
plt.step(t, x_quantized_2bit, where='mid', label='Two-bit Quantized')
plt.title('Two-bit Quantization')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()

print('Binary representation of one-bit quantized signal:', ' '.join(binary_1bit))
print('Binary representation of two-bit quantized signal:', ' '.join(binary_2bit))
```
